```python
import numpy as np
import matplotlib.pyplot as plt

list_mots_lowered_splited=[]
for mot in list_mots:
    mot=mot.lower()
    mot=list(mot)
    for lettre_in_mot in mot:
        list_mots_lowered_splited.append(lettre_in_mot)
list_lettres_count=[]
list_lettres_unique=list(dict.fromkeys(list_mots_lowered_splited))
for lettre in list_lettres_unique:
    list_lettres_count.append((lettre,list_mots_lowered_splited.count(lettre)))

print(list_lettres_count)
plt.hist(np.array(list_mots_lowered_splited), 70, facecolor='g', alpha=0.75)

plt.grid(True)
plt.show()    
    
    
    
    
```

```python
def plotfigure(dict):
    list_labels=[elt for elt in dict]
    plt.figure(figsize=(9, 3))
    for elt in list_labels:
        list=dict[elt]
        X=[]
        Y=[]
        for elmt in list:
            X.append(elmt[0])
            Y.append(elmt[1])
        plt.plot(X, Y)
    plt.ylim(0,70)
    plt.legend()
    plt.show()

plotfigure(dict)
            
            
    
```
