```python
n=int(input('enter un entier :'))
def factoriel(n):
    if n==0:
        return 1
    else :
        return n*factoriel((n-1))
        
factoriel(n)
```

```python
n=int(input('enter un entier :'))
x=float(input('enter x : '))
def puissance(n,x):
    if n==0:
        return 1
    else : 
        return x*puissance(n-1,x)

puissance(n,x)
```

```python
import numpy as np

n=int(input('enter un entier :'))
def jeudame(n):
    if n==0:
        return ('enter un entier strictement positive')
    else:
        A=np.full((n,n),'O')
        for i in range(n):
            A[i][i]='X'
        return A

jeudame(n)
    
    
```
